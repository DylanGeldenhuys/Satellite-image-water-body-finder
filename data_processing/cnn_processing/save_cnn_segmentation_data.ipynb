```python
import os
from pathlib import Path
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import reshape_as_image
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

def create_window(data, point, size):
    offset = int(size/2)
    window = data[point[0]-offset: point[0] + offset + 1,
                  point[1]-offset: point[1] + offset + 1]
    return(window)


image_data_dir = Path('/media/ds/New Volume/Waterbody_Project/Tiffs')

label_masks_dir = Path('/media/ds/New Volume/Waterbody_Project/labels')

segment_label_dir = Path('/media/ds/New Volume/Waterbody_Project/CNN_data/segment_label')
segment_input_dir = Path('/media/ds/New Volume/Waterbody_Project/CNN_data/segment_input')

```

```python
# must be odd number
data_resolution = 512

# size of largest window used in feature extraction, must be odd number
data_padding = 512

length = len(os.listdir(image_data_dir))
percentage_complete = 1
```

```python
for filename in os.listdir(image_data_dir):
    counter = 0
    # load files
    raster_image_data = rasterio.open(
        image_data_dir.joinpath(filename)).read()
    image_data = reshape_as_image(raster_image_data)

    mask_data = np.load(label_masks_dir.joinpath(
        filename.replace("tif", "npy")))

    # loop through image data and create training
    height = int(
        image_data.shape[0] / data_resolution - data_padding / data_resolution)
    width = int(image_data.shape[1] /
                data_resolution - data_padding / data_resolution)

    data_set = []

    offset = round(data_padding/2) + 100

    percentage_tracker = 10
    print('start')
    for j in range(height ):
        for i in range(width ):
            y = j * data_resolution + offset
            x = i * data_resolution + offset
            counter +=1
            #print(x,y)

            if (((j / height) * 100) > percentage_tracker):
                print("File: {0} {1}%".format(filename, percentage_tracker))
                percentage_tracker += 10

            # extract label
            label_window = create_window(mask_data, [y, x], data_padding)
            #print(label_window)
            #label = int(label_window.mean(axis=0).mean(axis=0) < 0.5)
            #if label == 1:
            #    counter += 1
                #print('1')
            sample_window = create_window(image_data, [y,x], data_padding)
            #print(sample_window.shape)
            #plt.savefig(segment_input_dir + '/'+ filename[:-4] + '{}'.format(counter) + '.png', sample_window)
            #plt.savefig(segment_label_dir + '/' +filename[:-4] + '{}'.format(counter) + '.png', label_window)
            #label_window = Image.fromarray(np.uint8(sample_window))
            #label_window.show()
            sample_window = Image.fromarray(np.uint8(sample_window))
            #print(sample_window)
            np.save(segment_label_dir.joinpath(filename[:-4] + '{}'.format(counter)),sample_window)
            #label_window.save(segment_label_dir.joinpath(filename[:-4] + '{}'.format(counter) + '.png'))
            sample_window.save(segment_input_dir.joinpath(filename[:-4] + '{}'.format(counter) + '.png'))

            # save sample window in label folder

            #elif label == 0:
            #    print('0')
            #    sample_window.save(no_water_body_dir.joinpath(filename[:-4] + "{}".format(counter)  + '.jpg'))

            

    print('\n')
    print("{}% total completed".format((percentage_complete / length * 100)))
    percentage_complete += 1


```
